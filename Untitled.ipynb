```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# Параметры модели
# -----------------------------
n_depth = 500
lithology_states = ['sand', 'shale', 'carbonate_sand', 'coal', 'siltstone']

# Диапазоны частоты встречаемости
lithology_freq_range = {
    'sand': (0.1, 0.25),
    'shale': (0.3, 0.55),
    'carbonate_sand': (0.05, 0.12),
    'coal': (0.01, 0.08),
    'siltstone': (0.1, 0.25)
}

# Диапазоны длины серии (количество точек)
series_length_range = {
    'sand': (2, 10),
    'shale': (1, 25),
    'carbonate_sand': (2, 6),
    'coal': (1, 4),
    'siltstone': (2, 12)
}

# Матрица переходов (Марковская цепь) для 5 литотипов
transition_matrix = np.array([
    [0.55, 0.2, 0.15, 0.05, 0.05],   # sand
    [0.25, 0.45, 0.1, 0.1, 0.1],     # shale
    [0.2, 0.1, 0.5, 0.1, 0.1],       # carbonate_sand
    [0.1, 0.1, 0.1, 0.6, 0.1],       # coal
    [0.2, 0.3, 0.1, 0.05, 0.35]      # siltstone
])

# -----------------------------
# Генерация начальной вероятности
# -----------------------------
def sample_initial_prob(freq_range):
    freqs = [np.random.uniform(low, high) for (low, high) in freq_range.values()]
    freqs = np.array(freqs)
    freqs /= freqs.sum()
    return freqs

initial_prob = sample_initial_prob(lithology_freq_range)

# -----------------------------
# Генерация литологии с сериями
# -----------------------------
def generate_lithology_series(n, states, trans_matrix, initial_prob, series_range):
    lithology = []
    while len(lithology) < n:
        if len(lithology) == 0:
            current_lith = np.random.choice(states, p=initial_prob)
        else:
            prev_lith_idx = states.index(current_lith)
            current_lith = np.random.choice(states, p=trans_matrix[prev_lith_idx])
        # длина серии
        min_len, max_len = series_range[current_lith]
        series_len = np.random.randint(min_len, max_len+1)
        lithology.extend([current_lith]*series_len)
    return lithology[:n]

lithology = generate_lithology_series(
    n_depth, lithology_states, transition_matrix, initial_prob, series_length_range
)

# -----------------------------
# Пористость и насыщение
# -----------------------------
phi, sw = [], []

for lith in lithology:
    if lith == 'sand':
        phi.append(np.random.normal(0.2, 0.03))
        sw.append(np.random.normal(0.3, 0.05))
    elif lith == 'shale':
        phi.append(np.random.normal(0.18, 0.02))
        sw.append(np.random.normal(0.8, 0.05))
    elif lith == 'carbonate_sand':
        phi.append(np.random.normal(0.12, 0.02))
        sw.append(np.random.normal(0.4, 0.05))
    elif lith == 'coal':
        phi.append(np.random.normal(0.15, 0.01))
        sw.append(np.random.normal(0.2, 0.05))
    elif lith == 'siltstone':
        phi.append(np.random.normal(0.11, 0.03))
        sw.append(np.random.normal(0.6, 0.1))

# -----------------------------
# Расчет ГК и плотности (ГГКП)
# -----------------------------
matrix_density = {
    'sand': 2.65,
    'shale': 2.55,
    'carbonate_sand': 2.71,
    'coal': 1.4,
    'siltstone': 2.62
}
rho_fluid = 1.0
gamma_base = {
    'sand': 40,
    'shale': 70,
    'carbonate_sand': 55,
    'coal': 30,
    'siltstone': 60
}

gamma_log, rho_log = [], []

for i, lith in enumerate(lithology):
    gamma_log.append(np.random.normal(gamma_base[lith], 3))
    rho = phi[i]*rho_fluid + (1-phi[i])*matrix_density[lith]
    rho += np.random.normal(0, 0.02)
    rho_log.append(rho)

# -----------------------------
# DataFrame
# -----------------------------
df = pd.DataFrame({
    'Depth': np.arange(n_depth),
    'Lithology': lithology,
    'Porosity': phi,
    'Sw': sw,
    'Gamma': gamma_log,
    'Rho': rho_log
})

# -----------------------------
# Визуализация каротажа
# -----------------------------
fig, axs = plt.subplots(1,3,figsize=(12,6), sharey=True)

axs[0].plot(df['Gamma'], df['Depth'], color='green')
axs[0].set_title('ГК')
axs[0].invert_yaxis()

axs[1].plot(df['Rho'], df['Depth'], color='blue')
axs[1].set_title('ГГКП (Density)')
axs[1].invert_yaxis()

axs[2].plot(df['Porosity'], df['Depth'], color='red')
axs[2].set_title('Пористость')
axs[2].invert_yaxis()

plt.tight_layout()
plt.show()

# -----------------------------
# Кроссплоты
# -----------------------------
palette_lith = {
    'sand': 'yellow',
    'shale': 'brown',
    'carbonate_sand': 'orange',
    'coal': 'black',
    'siltstone': 'gray'
}

# 1. Gamma vs Rho по литологии
plt.figure(figsize=(6,6))
sns.scatterplot(data=df, x='Gamma', y='Rho', hue='Lithology',
                palette=palette_lith, alpha=0.7)
plt.title('Crossplot Gamma vs Density по литологии')
plt.xlabel('ГК')
plt.ylabel('ГГКП (Density)')
plt.show()

# 2. Gamma vs Rho по пористости
plt.figure(figsize=(6,6))
sc = plt.scatter(df['Gamma'], df['Rho'], c=df['Porosity'],
                 cmap='viridis', alpha=0.7)
plt.colorbar(sc, label='Пористость')
plt.title('Crossplot Gamma vs Density по пористости')
plt.xlabel('ГК')
plt.ylabel('ГГКП (Density)')
plt.show()

# 3. Rho vs Porosity по литологии
plt.figure(figsize=(6,6))
sns.scatterplot(data=df, x='Rho', y='Porosity', hue='Lithology',
                palette=palette_lith, alpha=0.7)
plt.title('Crossplot Density vs Porosity по литологии')
plt.xlabel('ГГКП (Density)')
plt.ylabel('Пористость')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# Параметры модели
# -----------------------------
n_depth = 1000
lithology_states = ['sand', 'shale', 'carbonate_sand', 'coal', 'siltstone']

# Диапазоны частоты встречаемости
lithology_freq_range = {
    'sand': (0.1, 0.25),
    'shale': (0.35, 0.55),
    'carbonate_sand': (0.05, 0.12),
    'coal': (0.01, 0.08),
    'siltstone': (0.05, 0.12)
}

# Диапазоны длины серии (количество точек)
series_length_range = {
    'sand': (2, 10),
    'shale': (1, 25),
    'carbonate_sand': (2, 6),
    'coal': (1, 4),
    'siltstone': (2, 12)
}

# Матрица переходов (Марковская цепь)
transition_matrix = np.array([
    [0.6, 0.2, 0.1, 0.05, 0.05],
    [0.3, 0.5, 0.05, 0.05, 0.1],
    [0.2, 0.1, 0.5, 0.05, 0.15],
    [0.1, 0.1, 0.05, 0.7, 0.05],
    [0.2, 0.2, 0.2, 0.05, 0.35]
])

# -----------------------------
# Генерация начальной вероятности
# -----------------------------
def sample_initial_prob(freq_range):
    freqs = [np.random.uniform(low, high) for (low, high) in freq_range.values()]
    freqs = np.array(freqs)
    freqs /= freqs.sum()
    return freqs

initial_prob = sample_initial_prob(lithology_freq_range)

# -----------------------------
# Генерация литологии с сериями
# -----------------------------
def generate_lithology_series(n, states, trans_matrix, initial_prob, series_range):
    lithology = []
    while len(lithology) < n:
        if len(lithology) == 0:
            current_lith = np.random.choice(states, p=initial_prob)
        else:
            prev_lith_idx = states.index(current_lith)
            current_lith = np.random.choice(states, p=trans_matrix[prev_lith_idx])
        min_len, max_len = series_range[current_lith]
        series_len = np.random.randint(min_len, max_len+1)
        lithology.extend([current_lith]*series_len)
    return lithology[:n]

lithology = generate_lithology_series(n_depth, lithology_states, transition_matrix, initial_prob, series_length_range)

# -----------------------------
# Пористость и насыщение
# -----------------------------
phi, sw = [], []
for lith in lithology:
    if lith == 'sand':
        phi.append(np.random.normal(0.25, 0.03)); sw.append(np.random.normal(0.3, 0.05))
    elif lith == 'shale':
        phi.append(np.random.normal(0.15, 0.02)); sw.append(np.random.normal(0.8, 0.05))
    elif lith == 'carbonate_sand':
        phi.append(np.random.normal(0.20, 0.02)); sw.append(np.random.normal(0.4, 0.05))
    elif lith == 'coal':
        phi.append(np.random.normal(0.35, 0.04)); sw.append(np.random.normal(0.2, 0.05))
    elif lith == 'siltstone':
        phi.append(np.random.normal(0.18, 0.02)); sw.append(np.random.normal(0.6, 0.05))

# -----------------------------
# Базовые параметры ГК и плотности
# -----------------------------
matrix_density = {
    'sand':2.65, 'shale':2.55, 'carbonate_sand':2.71,
    'coal':1.4, 'siltstone':2.62
}
rho_fluid = 1.0
gamma_base = {
    'sand':40, 'shale':70, 'carbonate_sand':55,
    'coal':30, 'siltstone':60
}

# -----------------------------
# Генерация "чистых" каротажных кривых
# -----------------------------
gamma_log, rho_log = [], []
for i, lith in enumerate(lithology):
    gamma_log.append(np.random.normal(gamma_base[lith], 3))
    rho = phi[i]*rho_fluid + (1-phi[i])*matrix_density[lith]
    rho += np.random.normal(0, 0.02)
    rho_log.append(rho)

# -----------------------------
# Эффект инверсии (смешивание по краям слоев)
# -----------------------------
def apply_boundary_inversion(curve, lithology, window=3, strength=0.6):
    curve_new = curve.copy()
    n = len(curve)
    for i in range(1, n-1):
        if lithology[i] != lithology[i-1] or lithology[i] != lithology[i+1]:
            left = max(0, i-window)
            right = min(n, i+window+1)
            neighbor_mean = np.mean(curve[left:right])
            curve_new[i] = (1-strength)*curve[i] + strength*neighbor_mean
    return curve_new

gamma_inv = apply_boundary_inversion(np.array(gamma_log),  lithology, window=5)
rho_inv   = apply_boundary_inversion(np.array(rho_log),  lithology,  window=3)

# -----------------------------
# DataFrame
# -----------------------------
df = pd.DataFrame({
    'Depth': np.arange(n_depth),
    'Lithology': lithology,
    'Porosity': phi,
    'Sw': sw,
    'Gamma_clean': gamma_log,
    'Rho_clean': rho_log,
    'Gamma_inv': gamma_inv,
    'Rho_inv': rho_inv
})

# -----------------------------
# Визуализация кривых
# -----------------------------
fig, axs = plt.subplots(1,2,figsize=(12,6))

axs[0].plot(df['Gamma_clean'], df['Depth'], label='Чистый ГК', color='green', alpha=0.5)
axs[0].plot(df['Gamma_inv'], df['Depth'], label='ГК с инверсией', color='black')
axs[0].invert_yaxis(); axs[0].legend(); axs[0].set_title('Гамма-каротаж')

axs[1].plot(df['Rho_clean'], df['Depth'], label='Чистая плотность', color='blue', alpha=0.5)
axs[1].plot(df['Rho_inv'], df['Depth'], label='Плотность с инверсией', color='red')
axs[1].invert_yaxis(); axs[1].legend(); axs[1].set_title('ГГКП (Density)')

plt.show()

# -----------------------------
# Кроссплоты (с инверсией)
# -----------------------------
palette_lith = {
    'sand':'yellow','shale':'brown',
    'carbonate_sand':'orange','coal':'black',
    'siltstone':'gray'
}

plt.figure(figsize=(6,6))
sns.scatterplot(data=df, x='Gamma_inv', y='Rho_inv', hue='Lithology',
                palette=palette_lith, alpha=0.7)
plt.title('Crossplot Gamma vs Density (с инверсией)')
plt.xlabel('ГК'); plt.ylabel('ГГКП (Density)')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# Параметры модели
# -----------------------------
n_depth = 1000
depth_start, depth_end = 1000, 2000
depths = np.linspace(depth_start, depth_end, n_depth)

lithology_states = ['sand', 'shale', 'carbonate_sand', 'coal', 'siltstone']

# Частота встречаемости (непрямо влияет на начальные вероятности)
lithology_freq_range = {
    'sand': (0.01, 0.15),
    'shale': (0.35, 0.9),
    'carbonate_sand': (0.05, 0.12),
    'coal': (0.01, 0.08),
    'siltstone': (0.05, 0.15)
}

# Диапазоны длины серии
series_length_range = {
    'sand': (2, 10),
    'shale': (1, 155),
    'carbonate_sand': (2, 6),
    'coal': (1, 4),
    'siltstone': (2, 20)
}

# Базовая матрица переходов (Марковская цепь)
transition_matrix = np.array([
    [0.6, 0.2, 0.1, 0.05, 0.05],   # sand
    [0.3, 0.5, 0.1, 0.05, 0.05],   # shale
    [0.2, 0.1, 0.6, 0.05, 0.05],   # carbonate_sand
    [0.1, 0.1, 0.1, 0.6, 0.1],     # coal
    [0.2, 0.2, 0.2, 0.05, 0.35]    # siltstone
])

# -----------------------------
# Генерация начальных вероятностей
# -----------------------------
def sample_initial_prob(freq_range):
    freqs = [np.random.uniform(low, high) for (low, high) in freq_range.values()]
    freqs = np.array(freqs)
    freqs /= freqs.sum()
    return freqs

initial_prob = sample_initial_prob(lithology_freq_range)

# -----------------------------
# Генерация литологии: Марков + циклы трансгрессии/регрессии
# -----------------------------
def generate_lithology_markov_cycles(n, states, trans_matrix, initial_prob, series_range, cycle_len=200):
    lithology = []
    current_lith = np.random.choice(states, p=initial_prob)
    lithology.append(current_lith)

    for i in range(1, n):
        cycle_phase = (i // cycle_len) % 2  # 0 = трансгрессия, 1 = регрессия

        if cycle_phase == 0:  # Трансгрессия: sand→siltstone→shale
            bias_matrix = np.array([
                [0.4, 0.2, 0.05, 0.05, 0.3],
                [0.2, 0.4, 0.1, 0.05, 0.25],
                [0.1, 0.2, 0.5, 0.05, 0.15],
                [0.1, 0.1, 0.05, 0.7, 0.05],
                [0.15, 0.25, 0.2, 0.05, 0.35]
            ])
        else:  # Регрессия: shale→siltstone→sand
            bias_matrix = np.array([
                [0.6, 0.1, 0.1, 0.05, 0.15],
                [0.3, 0.3, 0.05, 0.05, 0.3],
                [0.3, 0.1, 0.4, 0.05, 0.15],
                [0.1, 0.1, 0.1, 0.6, 0.1],
                [0.3, 0.2, 0.1, 0.05, 0.35]
            ])

        prev_idx = states.index(current_lith)
        probs = 0.5 * trans_matrix[prev_idx] + 0.5 * bias_matrix[prev_idx]
        probs /= probs.sum()
        current_lith = np.random.choice(states, p=probs)
        lithology.append(current_lith)

    # Растягиваем серии
    lith_expanded = []
    for lith in lithology:
        min_len, max_len = series_range[lith]
        run_len = np.random.randint(min_len, max_len + 1)
        lith_expanded.extend([lith] * run_len)

    return lith_expanded[:n]

# -----------------------------
# Генерация литологии
# -----------------------------
lithology = generate_lithology_markov_cycles(
    n_depth, lithology_states, transition_matrix, initial_prob, series_length_range, cycle_len=200
)

# -----------------------------
# Свойства литологий
# -----------------------------
phi, sw = [], []
for lith in lithology:
    if lith == 'sand':
        phi.append(np.random.normal(0.25, 0.03))
        sw.append(np.random.normal(0.3, 0.05))
    elif lith == 'shale':
        phi.append(np.random.normal(0.15, 0.02))
        sw.append(np.random.normal(0.8, 0.05))
    elif lith == 'carbonate_sand':
        phi.append(np.random.normal(0.20, 0.02))
        sw.append(np.random.normal(0.4, 0.05))
    elif lith == 'coal':
        phi.append(np.random.normal(0.35, 0.04))
        sw.append(np.random.normal(0.2, 0.05))
    elif lith == 'siltstone':
        phi.append(np.random.normal(0.18, 0.02))
        sw.append(np.random.normal(0.5, 0.05))

# -----------------------------
# Каротажные кривые
# -----------------------------
matrix_density = {'sand':2.65, 'shale':2.55, 'carbonate_sand':2.71, 'coal':1.4, 'siltstone':2.62}
rho_fluid = 1.0
gamma_base = {'sand':40, 'shale':70, 'carbonate_sand':35, 'coal':30, 'siltstone':60}

gamma_log, rho_log = [], []
for i, lith in enumerate(lithology):
    gamma_log.append(np.random.normal(gamma_base[lith], 3))
    rho = phi[i]*rho_fluid + (1-phi[i])*matrix_density[lith]
    rho += np.random.normal(0, 0.02)
    rho_log.append(rho)

# -----------------------------
# Эффект инверсии (смазывание)
# -----------------------------
def apply_invasion_effect(curve, window=5, alpha=0.5):
    curve = np.array(curve)           # <- добавили
    kernel = np.ones(window) / window
    smooth = np.convolve(curve, kernel, mode="same")
    return alpha*smooth + (1-alpha)*curve

gamma_inv = apply_invasion_effect(gamma_log, window=7, alpha=0.6)
rho_inv = apply_invasion_effect(rho_log, window=7, alpha=0.6)

# -----------------------------
# DataFrame
# -----------------------------
df = pd.DataFrame({
    'Depth': depths,
    'Lithology': lithology,
    'Porosity': phi,
    'Sw': sw,
    'Gamma_clean': gamma_log,
    'Gamma_inv': gamma_inv,
    'Rho_clean': rho_log,
    'Rho_inv': rho_inv
})

# -----------------------------
# Визуализация каротажа
# -----------------------------
fig, axs = plt.subplots(1,3,figsize=(14,8), sharey=True)

axs[0].plot(df['Gamma_clean'], df['Depth'], label='Чистый ГК', color='green', alpha=0.5)
axs[0].plot(df['Gamma_inv'], df['Depth'], label='ГК с инверсией', color='black')
axs[0].set_title('ГК')
axs[0].set_ylim(depth_end, depth_start)
axs[0].legend()

axs[1].plot(df['Rho_clean'], df['Depth'], label='Чистая плотность', color='blue', alpha=0.5)
axs[1].plot(df['Rho_inv'], df['Depth'], label='Плотность с инверсией', color='red')
axs[1].set_title('ГГКП (Density)')
axs[1].set_ylim(depth_end, depth_start)
axs[1].legend()

axs[2].plot(df['Porosity'], df['Depth'], color='orange')
axs[2].set_title('Пористость')
axs[2].set_ylim(depth_end, depth_start)

plt.tight_layout()
plt.show()

# -----------------------------
# Кроссплоты
# -----------------------------
palette_lith = {
    'sand':'yellow','shale':'brown','carbonate_sand':'orange','coal':'black','siltstone':'gray'
}

plt.figure(figsize=(6,6))
sns.scatterplot(data=df, x='Gamma_inv', y='Rho_inv', hue='Lithology', palette=palette_lith, alpha=0.7)
plt.title('Crossplot Gamma vs Density по литологии')
plt.xlabel('ГК (inv)')
plt.ylabel('ГГКП (inv)')
plt.show()

plt.figure(figsize=(6,6))
sc = plt.scatter(df['Gamma_inv'], df['Rho_inv'], c=df['Porosity'], cmap='viridis', alpha=0.7)
plt.colorbar(sc, label='Пористость')
plt.title('Crossplot Gamma vs Density по пористости')
plt.xlabel('ГК (inv)')
plt.ylabel('ГГКП (inv)')
plt.show()

plt.figure(figsize=(6,6))
sns.scatterplot(data=df, x='Rho_inv', y='Porosity', hue='Lithology', palette=palette_lith, alpha=0.7)
plt.title('Crossplot Density vs Porosity по литологии')
plt.xlabel('ГГКП (inv)')
plt.ylabel('Пористость')
plt.show()
```
